class Network(object):
    
    # list 'sizes' contains number of neurons in respective layers
    def __init__(self, sizes):
        self.num_layers = len(sizes)
        self.sizes = sizes
        # first layer of neurons is an input layer without biases, which is why we use [1:]
        # weights stored as lists of Numpy matrices, so net.weights[1] connects second and third layer neurons
        self.biases = [np.random.rand(y, 1) for y in sizes[1:]]
        # since the weights will left multiply the inputs
        # the number of rows of one input layer equals the number of columns of the following matrix of weights
        # and the number of rows of the output layer equals the number of rows in that matrix of weights
        self.weights = [np.random.randn(y, x) for x, y in zip(sizes[:-1], sizes[1:])]
    
    def sigmoid(z):
        return 1.0/(1.0+np.exp(-z))
    
    def feedforward(self, a):
        for b, w in zip(self.biases, self.weights):
            a = sigmoid(np.dot(w, a)+b)
        return a
